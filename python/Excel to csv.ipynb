# Convert Excel file to csv files

```python
from pathlib import Path
import pandas as pd

INPUT_FILE = Path("/Users/anushanagula/Documents/tungsten/Data/ raw/project dataset.xlsx")
OUTPUT_DIR = Path("/Users/anushanagula/Documents/tungsten/Data/processed")

def convert_excel_to_csv(input_file: Path, output_dir: Path):
    output_dir.mkdir(parents=True, exist_ok=True)

    sheets = pd.read_excel(input_file, sheet_name=None)

    for sheet_name, df in sheets.items():
        clean_name = sheet_name.lower().replace(" ", "_")
        output_file = output_dir / f"{clean_name}.csv"

        df.to_csv(output_file, index=False)
        print(f"âœ… Saved {output_file}")


if __name__ == "__main__":
    convert_excel_to_csv(INPUT_FILE, OUTPUT_DIR)
```
